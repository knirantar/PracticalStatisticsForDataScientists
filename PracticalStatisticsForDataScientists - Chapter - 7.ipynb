<h2>Principal Component Analysis</h2>
Principal components analysis (PCA) is a technique to discover the way in which numeric variables covary.

The idea in PCA is to combine multiple numeric predictor variables into a smaller set of variables. the principal components, “explains” most of the variability of the full set of variables, reducing the dimension of the data.
